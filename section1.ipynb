```python
for i in range(len(df3)):
    event = df3.loc[i, 'event_type']
    cat = df.loc[i, 'category_code']
    if event == 'view':
        d[cat][0] += 1
    if event == 'purchase':
        d[cat][1] += 1
```

<h2>5.	What is top pair of categories which have high affinity for each other.<h2>

<h5>In here, I defined the the category affinity as 'A visitor’s activity in one session, such as which category they view most often, can be used for targeting in subsequent visits'.Each category page a visitor views during a session is captured, and his or her “favorite” category is calculated based on a recency and frequency model.
    Thus, I will caculate the cart/ view for each categories and extract the top pair of categories.<h5>

```python
d2 = collections.defaultdict(list)
df4 = df[['event_type', 'category_code']]
df4
```

```python
for i in range(len(df4)):
    cat = df.loc[i, 'category_code']
    d2[cat] = [0,0,0] #[key : category_code, value :[view count, cart count, average  (cart/ view)]]
```

```python
for i in range(len(df4)):
    event = df4.loc[i, 'event_type']
    cat = df4.loc[i, 'category_code']
    if event == 'view':
        d2[cat][0] += 1
    if event == 'cart':
        d2[cat][1] += 1
```

```python
for cat, val in d2.items():
    d2[cat][2] = (val[1] / val[0]) * 100

    
```
